````python
def build_potential(file):
    import pandas as pd

    # === 1. 读取数据 ===
    df = pd.read_excel(file)
    df.columns = df.columns.str.strip()
    df['数值'] = pd.to_numeric(df['数值'], errors='coerce')

    # === 2. 筛选出“建设用地整治潜力”类别 ===
    df_build = df[df['类别'] == '建设用地整治潜力'].copy()

    # === 3. 定义目标潜力项 ===
    target_items = [
        '低效工业用地腾退潜力',
        '存量低效用地潜力',
        '低效用地再开发潜力',
        '三旧改造潜力',
        '城镇低效用地再开发潜力'
    ]

    df_build = df_build[df_build['潜力项'].isin(target_items)]

    # === 4. 按地区取最大值 ===
    max_build = (
        df_build.loc[df_build.groupby('地区')['数值'].idxmax(),
                     ['地区', '潜力项', '数值']]
        .rename(columns={'潜力项': '最大潜力项', '数值': '建设用地整治潜力最大值'})
        .reset_index(drop=True)
    )
    data ={}
    data = max_build.set_index('地区')['建设用地整治潜力最大值'].to_dict()
    print(data)
    return data


file = '/mnt/e/共享文件夹/土地整治/data/潜力项数据_process.xlsx'
build_potent_1 = build_potential(file)
````

````python
import pandas as pd
import numpy as np
try:
    import altair as alt
except ImportError:
    alt = None

print("--- 开始执行【最终合理】分析 ---")



# --- 2. 数据准备 (分母 - 现状基底数据) ---

raw_table_data = """
| 地区 | 农用地 | 建设用地 | 生态保护 | 林地占比 |
| --- | --- | --- | --- | --- |
| 东莞凤岗 | 2833.345 | 562.2545 | 500.205 | 0.355758 |
| 东莞厚街 | 4038.608 | 453.5362 | 1137.282 | 0.251342 |
| 东莞塘厦 | 2818.813 | 405.0485 | 593.094 | 0.316887 |
| 东莞清溪 | 8233.237 | 5008.622 | 767.4774 | 0.410741 |
| 中山古镇 | 1158.895 | 3153.335 | 908.6506 | 0.008672 |
| 中山翠亨 | 9722.866 | 5454.885 | 7591.491 | 0.263625 |
| 云浮郁南 | 90859.71 | 4275.614 | 4514.456 | 0.705117 |
| 佛山三水 | 28619.79 | 25117.74 | 30952.13 | 0.178212 |
| 梅州兴宁 | 54803.89 | 6821.823 | 4330.765 | 0.629026 |
| 广州增城-简本 | 117899.3 | 29103.58 | 13394.62 | 0.417084 |
| 惠州大亚湾 | 16079.44 | 6036.813 | 941.9876 | 0.507028 |
| 惠州龙门 | 142236.4 | 8612.404 | 5821.635 | 0.799578 |
| 揭阳惠来（中部单元和东部单元） | 13682.37 | 3918.135 | 3738.856 | 0.223739 |
| 梅州大埔 | 211729.6 | 12151.79 | 6944.174 | 0.822717 |
| 梅州平远 | 101852.7 | 6538.384 | 2827.638 | 0.798181 |
| 梅州蕉岭 | 86065.76 | 7089.72 | 2980.527 | 0.79318 |
| 汕尾红海湾 | 6899.482 | 581.1954 | 1569.298 | 0.411746 |
| 汕尾陆丰 | 106045.1 | 14867.38 | 17850 | 0.373721 |
| 江门台山端芬 | 24601.76 | 1387.371 | 3944.303 | 0.621036 |
| 江门蓬江 | 8394.232 | 9816.686 | 5293.731 | 0.282054 |
| 江门鹤山（简版） | 61695.48 | 11142.02 | 12158.32 | 0.584944 |
| 江门鹤山（鹤城单元） | 12435.26 | 2103.378 | 1373.596 | 0.67737 |
| 河源紫金 | 273844.4 | 13071.35 | 5804.066 | 0.847446 |
| 清远佛冈（水头单元） | 13548.83 | 794.5712 | 277.2629 | 0.7898 |
| 清远英德（九龙单元） | 21836.74 | 1108.885 | 615.5623 | 0.684104 |
| 清远连南 | 30771.7 | 1455.98 | 655.0645 | 0.866499 |
| 湛江坡头 | 23578.11 | 7144.868 | 10791.63 | 0.187587 |
| 湛江廉江(2024-9) | 205917.5 | 39122.77 | 39629.42 | 0.365011 |
| 湛江廉江(2024-12) | 204619.5 | 39122.77 | 40927.51 | 0.360451 |
| 湛江徐闻曲界 | 25992.52 | 2186.18 | 689.8293 | 0.037752 |
| 湛江遂溪 | 86475.37 | 12747.11 | 6412.624 | 0.25443 |
| 湛江雷州龙门 | 32298.1 | 2993.576 | 1752.702 | 0.330785 |
| 肇庆鼎湖沙浦 | 6871.515 | 516.9161 | 4528.467 | 0.468156 |
| 茂名高州 | 279116.4 | 5575.716 | 19841.96 | 0.454455 |
| 阳江阳东红丰 | 8327.685 | 994.0908 | 1642.532 | 0.255799 |
| 韶关南雄 | 17432.11 | 1599.158 | 671.3879 | 0.693896 |
| 韶关始兴顿岗 | 7952.524 | 626.9087 | 435.208 | 0.558982 |
| 韶关曲江大塘 | 14398.3 | 1663.157 | 1067.709 | 0.633504 |
| 云浮云城 | 65589.84 | 9718.579 | 2502.253 | 0.687761 |
| 广州黄埔 | 22770.28 | 328.9905 | 1668.051 | 0.649602 |
| 韶关新丰马头 | 48731.82 | 1150.204 | 1035.896 | 0.887519 |
"""

# 这是您需要的“3类数据获取的代码”
denom_agri = {}
denom_build = {}
denom_eco = {}

# 解析上述表格字符串
for line in raw_table_data.strip().split('\n'):
    if line.startswith('| 地区') or line.startswith('| ---'):
        continue
    
    parts = [col.strip() for col in line.strip('|').split('|')]
    
    if len(parts) >= 4:
        region = parts[0]
        try:
            # 转换为浮点数
            denom_agri[region] = float(parts[1])
            denom_build[region] = float(parts[2])
            denom_eco[region] = float(parts[3])
        except ValueError:
            print(f"跳过无效行: {region}")

print("--- 步骤 2: 已成功解析 3 类现状基底(分母)数据 ---")
print(f"农用地 (分母) 示例: '东莞凤岗': {denom_agri.get('东莞凤岗')}")
print(f"建设用地 (分母) 示例: '东莞凤岗': {denom_build.get('东莞凤岗')}")
print(f"生态保护 (分母) 示例: '东莞凤岗': {denom_eco.get('东莞凤岗')}")


# --- 3. 解决地区不一致问题 ---
# "分子" (潜力数据) 使用了长而乱的名字
# "分母" (现状数据) 使用了短而净的名字
# 我们需要将 "分子" 的名字 映射到 "分母" 的名字

numerator_key_map = {
    # 乱名 (分子) : 净名 (分母)
    '中山古': '中山古镇',
    '中山翠亨新': '中山翠亨',
    '云浮郁南全域土地综整治（一期）': '云浮郁南',
    '梅州兴宁（黄槐等六单元）': '梅州兴宁',
    '兴宁（黄槐等六单元）': '梅州兴宁',
    '广州增城--简本': '广州增城-简本',
    '惠州大亚湾开发（西街道、澳头街道、霞涌街道单元）': '惠州大亚湾',
    '梅州平远（简版）': '梅州平远',
    '汕尾红海湾经济开发': '汕尾红海湾',
    '江门蓬江专家评审稿': '江门蓬江',
    '河源紫金无附件版': '河源紫金',
    '清远连南瑶族自治': '清远连南',
    '湛江坡头项目（坡头等4个单元）': '湛江坡头',
    '湛江廉江 2024年9月': '湛江廉江(2024-9)',
    '湛江廉江2024年12月': '湛江廉江(2024-12)',
    '湛江徐闻项目（曲界单元）': '湛江徐闻曲界',
    '湛江遂溪（西片单元）': '湛江遂溪',
    '湛江雷州（龙门单元）': '湛江雷州龙门',
    '肇庆鼎湖（沙浦单元）': '肇庆鼎湖沙浦',
    '阳江阳东（红丰单元） - 简版': '阳江阳东红丰',
    '韶关始兴（顿岗单元）': '韶关始兴顿岗',
    '韶关曲江（大塘单元）': '韶关曲江大塘',
    '新丰马头': '韶关新丰马头',
    '东莞凤岗项目': '东莞凤岗' # 这个在 build_potent_1 等字典中
}

def remap_keys(d, key_map):
    """递归地重映射字典或字典列表中的键"""
    if not isinstance(d, dict):
        return d
    return {key_map.get(k, k): v for k, v in d.items()}

# 3.1 清理 7 个"分子"字典的键
farm_potent_clean = remap_keys(farm_potent, numerator_key_map)
farm_potent_1_clean = remap_keys(farm_potent_1, numerator_key_map)
build_potent_1_clean = remap_keys(build_potent_1, numerator_key_map)
build_potent_2_clean = remap_keys(build_potent_2, numerator_key_map)
ecological_potent_1_clean = remap_keys(ecological_potent_1, numerator_key_map)
ecological_potent_2_clean = remap_keys(ecological_potent_2, numerator_key_map)
ecological_potent_3_clean = remap_keys(ecological_potent_3, numerator_key_map)

print("--- 步骤 3: 已解决地区名称不一致问题 ---")

# --- 4. 整合数据 (DataFrame) ---

# 4.1. "分子" DataFrame (绝对值)
data_dict = {
    '新增耕地潜力': farm_potent_clean,
    '高标准农田建设潜力': farm_potent_1_clean,
    '低效用地潜力': build_potent_1_clean,
    '建设用地增减挂钩(拆旧复垦)潜力': build_potent_2_clean,
    '流域生态修复潜力': ecological_potent_1_clean,
    '造林绿化潜力': ecological_potent_2_clean,
    '矿山修复潜力': ecological_potent_3_clean
}
df_potential = pd.DataFrame(data_dict)

# 4.2. "分母" Series (现状基底)
s_denom_agri = pd.Series(denom_agri, name="现状农用地")
s_denom_build = pd.Series(denom_build, name="现状建设用地")
s_denom_eco = pd.Series(denom_eco, name="现状生态保护")

# --- 5. 对齐和预处理 ---
# 合并所有分母为一个DataFrame
df_denominators = pd.DataFrame({
    'agri_denom': s_denom_agri,
    'build_denom': s_denom_build,
    'eco_denom': s_denom_eco
})

# 关键: 使用 'inner' join, 确保只保留同时具有 分子(潜力) 和 分母(现状) 数据的地区
df_potential_aligned, df_denominators_aligned = df_potential.align(
    df_denominators, join='inner', axis=0
)

# 5.1. 填充 "分子" (绝对值) 的 NaN 为 0
df_potential_filled = df_potential_aligned.fillna(0)

# 5.2. 准备 "分母" (避免除以0)
df_denominators_aligned = df_denominators_aligned.replace(0, np.nan)

print(f"--- 步骤 5: 数据对齐：成功对齐 {len(df_potential_aligned)} 个地区 ---")

# --- 6. 【!! 核心逻辑 !!】: 计算“真实强度”矩阵 (df_ratios) ---
# (潜力 / 现状基底)
df_ratios = pd.DataFrame(index=df_potential_aligned.index)

# 6.1. 农业潜力强度 (除以 农用地)
df_ratios['新增耕地潜力'] = df_potential_filled['新增耕地潜力'] / df_denominators_aligned['agri_denom']
df_ratios['高标准农田建设潜力'] = df_potential_filled['高标准农田建设潜力'] / df_denominators_aligned['agri_denom']

# 6.2. 建设潜力强度 (除以 建设用地)
df_ratios['低效用地潜力'] = df_potential_filled['低效用地潜力'] / df_denominators_aligned['build_denom']
df_ratios['建设用地增减挂钩(拆旧复垦)潜力'] = df_potential_filled['建设用地增减挂钩(拆旧复垦)潜力'] / df_denominators_aligned['build_denom']

# 6.3. 生态潜力强度 (全部除以 生态保护)
df_ratios['流域生态修复潜力'] = df_potential_filled['流域生态修复潜力'] / df_denominators_aligned['eco_denom']
df_ratios['造林绿化潜力'] = df_potential_filled['造林绿化潜力'] / df_denominators_aligned['eco_denom']
df_ratios['矿山修复潜力'] = df_potential_filled['矿山修复潜力'] / df_denominators_aligned['eco_denom']

# 6.4. 填充 0/0 或 x/NaN 导致的 NaN
df_ratios = df_ratios.fillna(0)
print("--- 步骤 6: 已计算【真实强度】矩阵 (df_ratios) ---")


# --- 7. 对“强度矩阵”进行分级 (1-5分) ---
df_scores = pd.DataFrame(index=df_ratios.index)
for column in df_ratios.columns:
    col_data = df_ratios[column]
    scores_series = pd.Series(1.0, index=col_data.index)
    non_zero_data = col_data[col_data > 0]
    
    if not non_zero_data.empty:
        try:
            bins = pd.qcut(non_zero_data, q=5, labels=False, duplicates='drop')
            scores = bins + 1
        except ValueError:
            ranks_pct = non_zero_data.rank(pct=True)
            scores = np.ceil(ranks_pct * 5)
            scores[scores == 0] = 1 
        scores_series.loc[non_zero_data.index] = scores
    df_scores[column] = scores_series.astype(int)
print("--- 步骤 7: 已完成 *强度*得分 (1-5分) 计算 ---")

# --- 8. 定义权重并计算加权得分 ---
weights = pd.Series({
    '新增耕地潜力': 0.2, '高标准农田建设潜力': 0.2, '低效用地潜力': 0.2,
    '建设用地增减挂钩(拆旧复垦)潜力': 0.2, '流域生态修复潜力': 0.2/3,
    '造林绿化潜力': 0.2/3, '矿山修复潜力': 0.2/3
}, name="Weights")
weights = weights[df_ratios.columns] 

df_weighted_scores = df_scores * weights

# --- 9. 按3个角度汇总得分 ---
df_angle_scores = pd.DataFrame(index=df_weighted_scores.index)
df_angle_scores['农业整治潜力得分'] = df_weighted_scores['新增耕地潜力'] + df_weighted_scores['高标准农田建设潜力']
df_angle_scores['建设用地整治潜力得分'] = df_weighted_scores['低效用地潜力'] + df_weighted_scores['建设用地增减挂钩(拆旧复垦)潜力']
df_angle_scores['生态修复潜力得分'] = df_weighted_scores['流域生态修复潜力'] + df_weighted_scores['造林绿化潜力'] + df_weighted_scores['矿山修复潜力']
df_angle_scores['最终综合潜力'] = df_angle_scores.sum(axis=1)
print("--- 步骤 9: 已按“农业、建设、生态”3个角度汇总得分 ---")

# --- 10. 整合最终结果 (包含3个角度和总分) ---
df_final_sorted = df_angle_scores.sort_values(by="最终综合潜力", ascending=False)

print("\n--- 最终综合得分 (按3个角度汇总) 排序结果 示例 (前10) ---")
print(df_final_sorted.head(10))

# --- 11. 保存到 CSV (包含3个角度和总分) ---
output_csv = "region_potential_FINAL_scores.csv"
df_final_sorted.to_csv(output_csv, index_label='地区名称', encoding='utf-8-sig')
print(f"\n处理完成！")
print(f"已保存包含[农业/建设/生态]得分的CSV: {output_csv}")
````

````python
import pandas as pd
import numpy as np



# data_dict={
#     '新增耕地潜力':farm_potent ,
#     '高标准农田建设潜力': farm_potent_1,
#     '低效用地潜力': build_potent_1 ,
#     '建设用地增减挂钩(拆旧复垦)潜力': build_potent_2,
#     '流域生态修复潜力': ecological_potent_1 ,
#     '造林绿化潜力': ecological_potent_2,
#     '矿山修复潜力': ecological_potent_3 
# }
# df_potential = pd.DataFrame(data_dict)

def remap_keys(d, key_map):
    """递归地重映射字典或字典列表中的键"""
    if not isinstance(d, dict):
        return d
    return {key_map.get(k, k): v for k, v in d.items()}

# 您的数据中存在名称不一致, 例如 '东莞凤岗' 和 '东莞凤岗项目'
# 我们在这里定义一个映射表, 将它们统一
key_map = {
    '东莞凤岗项目': '东莞凤岗'}

# 2.1. 清理7个潜力字典的键
farm_potent_clean = remap_keys(farm_potent, key_map)
farm_potent_1_clean = remap_keys(farm_potent_1, key_map)
build_potent_1_clean = remap_keys(build_potent_1, key_map)
build_potent_2_clean = remap_keys(build_potent_2, key_map)
ecological_potent_1_clean = remap_keys(ecological_potent_1, key_map)
ecological_potent_2_clean = remap_keys(ecological_potent_2, key_map)
ecological_potent_3_clean = remap_keys(ecological_potent_3, key_map)
# 2.2. 清理总潜力字典的键
potential_data_clean = remap_keys(potential_data, key_map)


# 2.3. 创建 "分子" DataFrame (绝对值)
data_dict = {
    '新增耕地潜力': farm_potent_clean,
    '高标准农田建设潜力': farm_potent_1_clean,
    '低效用地潜力': build_potent_1_clean,
    '建设用地增减挂钩(拆旧复垦)潜力': build_potent_2_clean,
    '流域生态修复潜力': ecological_potent_1_clean,
    '造林绿化潜力': ecological_potent_2_clean,
    '矿山修复潜力': ecological_potent_3_clean
}
df_potential = pd.DataFrame(data_dict)

# 2.4. 创建 "分母" Series (总潜力)
s_total_area = pd.Series(potential_data_clean, name="总潜力面积")

# --- 3. 对齐和预处理 ---

# 3.1. 对齐 "分子" 和 "分母"
# join='inner' 确保只保留在 *两个* 数据源中都存在的地区
df_potential_aligned, s_total_area_aligned = df_potential.align(
    s_total_area, join='inner', axis=0
)

# 3.2. 填充 "分子" (绝对值) 的 NaN 为 0
df_potential_filled = df_potential_aligned.fillna(0)

# 3.3. 准备 "分母"
# 避免除以0: 如果总潜力为0，则将其设为 NaN，这样除法会产生 NaN
s_total_area_aligned[s_total_area_aligned == 0] = np.nan

print(f"--- 数据对齐：成功对齐 {len(df_potential_aligned)} 个地区 ---")
if len(df_potential_aligned) == 0:
    print("错误：数据对齐后没有剩余地区。请检查地区名称是否完全不匹配。")
else:
    # --- 4. 计算“密度/占比”矩阵 (df_ratios) ---
    # 这是您新思路的核心：用“分子”除以“分母”
    df_ratios = df_potential_filled.div(s_total_area_aligned, axis=0)
    
    # 用 0 填充除法产生的 NaN (例如 x/NaN 或 0/0)
    df_ratios = df_ratios.fillna(0)
    
    print("--- 步骤 4: 已计算潜力密度/占比矩阵 (df_ratios) ---")
    # print(df_ratios.head()) # 取消注释以查看中间结果

    # --- 5. 对“密度矩阵”进行分级 (1-5分) ---
    # 规则:
    # - 值为 0 (包括NaN填充的) = 1分 (最低)
    # - 值 > 0, 根据其在非0数据中的分位数分为 1-5 级 (高值高分)
    df_scores = pd.DataFrame(index=df_ratios.index)

    for column in df_ratios.columns:
        col_data = df_ratios[column]
        
        # 默认所有地区得1分 (针对0值)
        scores_series = pd.Series(1.0, index=col_data.index)
        
        # 筛选出大于0的数据进行分级
        non_zero_data = col_data[col_data > 0]
        
        if not non_zero_data.empty:
            # 尝试使用 qcut (分位数) 分为5级
            try:
                # labels=False 返回 0-4 的 bin 索引
                # duplicates='drop' 处理重复的边界值
                bins = pd.qcut(non_zero_data, q=5, labels=False, duplicates='drop')
                
                # 将 0-4 的 bin 索引转换为 1-5 的得分
                scores = bins + 1
                
            except ValueError:
                # 如果 qcut 失败 (例如, 唯一值太少或数据太集中)
                # 我们退回到使用 rank (排名)
                # rank(pct=True) 将排名标准化到 0.0 - 1.0 之间
                ranks_pct = non_zero_data.rank(pct=True)
                
                # 将 0.0-1.0 的百分位排名 转换为 1-5 的得分
                scores = np.ceil(ranks_pct * 5)
                scores[scores == 0] = 1 # 确保最低分是1

            # 将计算出的 1-5 分 (仅非0数据) 更新回 scores_series
            scores_series.loc[non_zero_data.index] = scores
            
        df_scores[column] = scores_series.astype(int) # 存储为整数
    
    print("--- 步骤 5: 已完成 *密度*得分 (1-5分) 计算 ---")
    # print(df_scores.head()) # 取消注释以查看中间结果

    # --- 6. 加权计算最终得分 ---
    weights = pd.Series({
        '新增耕地潜力': 0.2,
        '高标准农田建设潜力': 0.2,
        '低效用地潜力': 0.2,
        '建设用地增减挂钩(拆旧复垦)潜力': 0.2,
        '流域生态修复潜力': 0.2 / 3,  # (0.0666... 保持精度)
        '造林绿化潜力': 0.2 / 3,
        '矿山修复潜力': 0.2 / 3
    }, name="Weights")
    
    # 确保权重的顺序和列与 df_ratios 一致
    weights = weights[df_ratios.columns] 

    # (1-5分) * (各自权重)
    df_weighted_scores = df_scores * weights
    # axis=1: 按行相加 (计算每个地区的总分)
    df_final_scores = df_weighted_scores.sum(axis=1)
    df_final_scores.name = "Final_Density_Score" # 最终得分列的名称

    # --- 7. 整合最终结果 ---
    # 将 原始绝对值、密度占比、1-5分、最终得分 合并
    
    # 为列名添加后缀以便区分
    df_final_results = df_potential_filled.add_suffix('_绝对值') \
                  .join(df_ratios.add_suffix('_占比')) \
                  .join(df_scores.add_suffix('_得分')) \
                  .join(df_final_scores)

    # 7.1. 排序
    df_final_sorted = df_final_results.sort_values(by="Final_Density_Score", ascending=False)
    
    print("--- 步骤 7: 已整合最终结果并排序 ---")
    
    # 打印前10行结果的示例
    print("\n--- 最终综合得分 (基于密度) 排序结果 示例 (前10) ---")
    display_cols = ['Final_Density_Score', '新增耕地潜力_占比', '新增耕地潜力_得分', '低效用地潜力_占比', '低效用地潜力_得分']
    print(df_final_sorted[display_cols].head(10))

    # --- 8. 保存到 CSV ---
    output_csv = "region_potential_density_scores.csv"
    df_final_sorted.to_csv(output_csv, index_label='地区名称', encoding='utf-8-sig')
    print(f"\n处理完成！")
    print(f"详细结果已保存到文件: {output_csv}")
    print("文件中包含了每个地区的 [绝对值], [占比], [1-5分得分], 以及 [最终综合得分]。")

````
